## Google trad paraphrases

```python
from googletrans import Translator
#Let us see what it gives out 
translator=Translator()
def paraphraser(query): 
    test= translator.translate(text=query, src="en", dest="it")
    temp= translator.translate(text=test.text, src="it", dest="ar")
    secondstep= translator.translate(text = temp.text, src="ar", dest="ja")
    fulltest= translator.translate(text=secondstep.text, src="ja", dest="en")
    return fulltest.text 
```
